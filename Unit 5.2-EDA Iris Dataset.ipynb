# Iris Flower dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#download iris.csv from https://archive.ics.uci.edu/ml/datasets/Iris'
# Load the csv file into a pandas Dataframe
iris = pd.read_csv("iris.csv")
```

```python
# Explore the number of features and data points
print(iris.shape)
```

```python
# what are the various column names in the dataset
print(iris.columns)
```

```python
# Question we need to identify is what are the data points for each class
# or How many flowers for each species are present

iris['species'].value_counts()
```

```python
# 2- D Scatter Plot
# understand the X and Y axis, Labels and Scale

iris.plot(kind='scatter', x = 'sepal_length', y = 'sepal_width');
plt.grid(True)
plt.show()
```

```python
# Plot the same scatter plot using SEABORN using colour plotting

sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", height = 6) \
    .map(plt.scatter, "sepal_length", "sepal_width") \
    .add_legend();
plt.show();
```
